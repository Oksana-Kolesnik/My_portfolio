```python
from scipy.stats import ttest_ind
```

```python
s, p = ttest_ind(df[df['id_group']==0]['nflag_order'],df[df['id_group']==1]['nflag_order'])
```

```python
for i in df['name_city'].unique():
    print(i)
    s,p = ttest_ind(df[(df['name_city']==i) & (df['id_group']==0)]['nflag_order'], df[(df['name_city']==i) & (df['id_group']==1)]['nflag_order'])
    print(p)
    print(s)
    print()
```

```python
df['hour'] = df['time_view'].dt.hour
```

```python
df['flag_hour'] = 0
```

```python
df_peak_orders = df[df['hour'].isin([7, 8, 9, 10, 17, 18, 19, 20])]
```

```python
for i in df['hour']:
    if i in df_peak_orders['hour'].values:
        df.loc[df['hour'] == i, 'flag_hour'] = 1
    else:
        df.loc[df['hour'] == i, 'flag_hour'] = 0
```

```python
df
```

```python
#час пик
s,p = ttest_ind(df[(df['flag_hour']==1) & (df['id_group']==0)]['nflag_order'], df[(df['flag_hour']==1) & (df['id_group']==1)]['nflag_order'])
print(p)
print(s)
```

```python
#обычное время
s,p = ttest_ind(df[(df['flag_hour']==0) & (df['id_group']==0)]['nflag_order'], df[(df['flag_hour']==0) & (df['id_group']==1)]['nflag_order'])
print(p)
print(s)
```
